    'nua_01z2': '주간 아침식사 일수', ## 최근 1년간 주당 아침식사 일수
                (주 5~7일 = 1, 주 3~4일 = 2, 주 1~2일 = 3, 주 0회 = 4, 응답거부 = 7, 모름 = 9)
    'oba_01z1': '체형(5단계)'

```python
import pandas as pd
import numpy as np

#미리 받은 원시자료를 excel에서 서울시 데이터만 따로 분리하여 추출하였습니다.
obesity = pd.read_csv('chs22_서울만.csv')
obesity.head()
```

### 분석에 필요한 칼럼만 추출

```python
obesity_df = obesity[['oba_02z1', 'oba_03z1','drb_01z3', 'drb_03z1', 'smb_01z1', 
                      'pha_04z1', 'pha_07z1', 'phb_01z1', 'nua_01z2', 'oba_01z1']]
obesity_df
```

### 결측치 확인

```python
obesity_df.isnull().sum()
```

### 칼럼명을 이해하기 쉽게 변경

```python
obesity_df = obesity_df.rename(columns = {
    'oba_02z1': '키', 
    'oba_03z1': '몸무게', 
    'pha_04z1': '격렬한 신체활동 일수', 
    'pha_07z1': '중증도 신체활동 일수', 
    'phb_01z1': '걷기 실천 일수',  
    'drb_01z3': '연간 음주 빈도', 
    'drb_03z1': '한번 섭취시 음주량', 
    'smb_01z1': '일반담배 하루 흡연량', 
    'nua_01z2': '주간 아침식사 일수',
    'oba_01z1': '체형(5단계)'
})
obesity_df
```

#### 변경한 Dataframe을 새로운 csv파일로 저장

```python
obesity_df.to_csv('obesity_df.csv', encoding = 'cp949')
```

```python
obesity_df6.describe()
```

### 스케일 맞추기(정규화)

주간 아침식사 일수의 범주가 다른 범주들과 반대로 라벨링이 되어 있으니 이에 대한 라벨링 재설정

(1 -> 3, 2 -> 2, 3 -> 1, 4 -> 0)

#### 주간 아침식사 일수의 라벨 재설정

```python
# '주간 아침식사 일수' 칼럼의 라벨을 변경하기 위한 딕셔너리
new_labels = {
    1: 3,
    2: 2,
    3: 1,
    4: 0
}

# replace 메서드를 사용하여 칼럼의 값 매핑
obesity_df6['주간 아침식사 일수'] = obesity_df6['주간 아침식사 일수'].replace(new_labels)
obesity_df6
```

```python
obesity_df6.describe()
```

```python
# obesity_df6을 obesity_df_prep라는 새로운 변수에 저장
obesity_df_prep = obesity_df6
obesity_df_prep
```

```python
obesity_df_prep.to_csv('obesity_df_prep.csv', encoding = 'cp949')
```

#### 각 칼럼들에 대하여 표준 정규화(z-score 이용)
$$ z = \frac{(X - m)}{\sigma} $$


```python
from sklearn.preprocessing import StandardScaler

# StandardScaler 객체 생성
scaler = StandardScaler()

# fit_transform 메서드를 사용하여 표준 정규화 수행
obesity_df_prep_std = pd.DataFrame(scaler.fit_transform(obesity_df_prep.iloc[:,:-1]), columns=obesity_df_prep.columns[:-1])
obesity_df_prep_std[obesity_df_prep.columns[-1]] = obesity_df_prep[obesity_df_prep.columns[-1]].values
obesity_df_prep_std.head()
```

```python
obesity_df_prep_std.to_csv('obesity_std.csv', encoding = 'cp949')
```

#### MinMax-Scaling을 이용한 정규화
$$ X_s = \frac{x - min}{\max-min} $$

```python
from sklearn.preprocessing import MinMaxScaler

mm_scaler = MinMaxScaler()

obesity_df_prep_mm = pd.DataFrame(mm_scaler.fit_transform(obesity_df_prep.iloc[:, :-1]), columns = obesity_df_prep.columns[:-1])
obesity_df_prep_mm[obesity_df_prep.columns[-1]] = obesity_df_prep[obesity_df_prep.columns[-1]].values
obesity_df_prep_mm.head()
```

```python
obesity_df_prep_mm.to_csv('obesity_minmax.csv', encoding = 'cp949')
```
